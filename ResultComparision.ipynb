### Rate Results

```python
csv_rate_files = load_files('./results/rate/', 'csv')
csv_rate_files
```

```python
def get_model_names(files):
    model_names = []
    for file in files:
        # Extract the model name from the file path
        model_name = file.split('/')[-1].replace('predictions_', '').replace('.csv', '')
        model_names.append(model_name)
    return model_names
```

```python
def find_highest_impact_results_df(files, limit):
    results_df =  pd.DataFrame()
    # Iterate over each CSV file
    for file in files:
        # Extract the model name from the file path
        model_name = file.split('/')[-1].replace('predictions_', '').replace('.csv', '')
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file)
        # code for plotting
        # create a 3D scatter plot for genres, cast, and directors
        fig = plt.figure(figsize=(10,10))
        ax = fig.add_subplot(111, projection='3d')
        # Set the limits for the x, y, and z axes
        ax.set_xlim(0, limit)
        ax.set_ylim(0, limit)
        ax.set_zlim(0, limit)
        # Iterate over each row in the DataFrame
        for index, row in df.iterrows():
            user_id = row['userId']
            # Check if the 'userId' column exists in the 'results_df' DataFrame
            if 'userId' not in results_df.columns:
                # Create a new 'userId' column with initial NaN values
                results_df['userId'] = user_id
            # Check if the userId already exists in the results_df DataFrame
            if user_id not in results_df['userId'].values:
                # Concatenate the new user DataFrame with the existing 'results_df' DataFrame
                results_df = pd.concat([results_df, pd.DataFrame({'userId': [user_id]})], ignore_index=True)
            
            # Add a new column to the results_df DataFrame with the highest impact dimension
            dim_column_name = f'highestImpact_{model_name}'
            highest_impact_dim = df[['mse_wo_genres', 'mse_wo_Cast', 'mse_wo_Directors']].idxmax(axis=1)[index]
            # Add a new column to the results_df DataFrame with the mse value
            mse_column_name = f'highestMSE_{model_name}'
            results_df.loc[results_df['userId'] == user_id, mse_column_name] = row[highest_impact_dim]
            results_df.loc[results_df['userId'] == user_id, dim_column_name] = highest_impact_dim.replace('mse_wo_', '')

            genre_score = row['mse_wo_genres']
            cast_score = row['mse_wo_Cast']
            director_score = row['mse_wo_Directors']
            # plot the point
            if highest_impact_dim == 'mse_wo_genres':
                genre_plot = ax.scatter(genre_score, cast_score, director_score, color='red', marker='o', alpha = 0.3)
            elif highest_impact_dim == 'mse_wo_Cast':
                cast_plot = ax.scatter(genre_score, cast_score, director_score, color='blue', marker='o', alpha = 0.5)
            else:
                dir_plot = ax.scatter(genre_score, cast_score, director_score, color='green', marker='o', alpha = 0.5)
      # annotate the point with the user ID
      # ax.text(genre_score, cast_score, director_score, row['userId'], color='black')
        # set the labels and title
        ax.set_xlabel('Genre Score', fontsize=20)
        ax.set_ylabel('Cast Score', fontsize=20)
        ax.set_zlabel('Director Score', fontsize=20)
        ax.set_title('Highest Impact identified by '+ model_name, fontsize=20)
        # add the legend
        try:
            ax.legend(handles = [genre_plot, dir_plot, cast_plot], labels=['Genres', 'Directors', 'Casts'], fontsize=20)
        except:
            ax.legend(handles = [genre_plot, dir_plot], labels=['Genres', 'Directors'], fontsize=20)

        plt.show()
    return results_df
```

```python
rate_results_df = find_highest_impact_results_df(csv_rate_files, 1.5)
rate_results_df
```

### Rate Visualizations

```python
# Define the model names
rate_model_names = get_model_names(csv_rate_files)
rate_model_names

# Define the dimensions
dimensions = ['genres', 'Cast', 'Directors']
```

```python
genre_counts = []
cast_counts = []
director_counts = []

for model_name in rate_model_names:
    highest_impact_col = f'highestImpact_{model_name}'
    genre_count = rate_results_df[highest_impact_col].value_counts().get('genres', 0)
    cast_count = rate_results_df[highest_impact_col].value_counts().get('Cast', 0)
    director_count = rate_results_df[highest_impact_col].value_counts().get('Directors', 0)
    genre_counts.append(genre_count)
    cast_counts.append(cast_count)
    director_counts.append(director_count)

genre_counts.append(baseline_df_top100['rate_affecting_dim'].value_counts().get('genres', 0))
cast_counts.append(baseline_df_top100['rate_affecting_dim'].value_counts().get('Cast', 0))
director_counts.append(baseline_df_top100['rate_affecting_dim'].value_counts().get('Directors', 0))

# Set the x-axis positions
x_pos = np.arange(len(rate_model_names)+1) 

# Set the width of each bar
bar_width = 0.2

# Create the bar plot
plt.bar(x_pos - bar_width, genre_counts, width=bar_width, label='Genre', color='red', alpha = 0.7)
plt.bar(x_pos, cast_counts, width=bar_width, label='Cast', color='blue', alpha = 0.7)
plt.bar(x_pos + bar_width, director_counts, width=bar_width, label='Director', color='green', alpha = 0.7)

# Set the x-axis labels
plt.xticks(x_pos, rate_model_names + ['Baseline'], rotation=90)

# Set the y-axis label
plt.ylabel('Count')

# Set the plot title
plt.title('Counts of Genre, Cast, and Director as \n Rate affecting dimensions for various models')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Initialize a dictionary to store the counts for each model
comparison_counts = {}

# Iterate over each model
for model_name in rate_model_names:
    # Initialize the count to 0 for the current model
    comparison_counts[model_name] = 0
    
    # Iterate over each user in the baseline dataframe
    for user_id in baseline_df_top100['userId']:
        # Find the corresponding row in the rate_results_df dataframe
        user_row = rate_results_df.loc[rate_results_df['userId'] == user_id]
        
        # Check if the highest impact dimension in rate_results_df is the same as the rate-affecting dimension in the baseline dataframe
        if user_row['highestImpact_' + model_name].values[0] == baseline_df_top100.loc[baseline_df_top100['userId'] == user_id, 'rate_affecting_dim'].values[0]:
            # Increment the count if they are the same
            comparison_counts[model_name] += 1

# Add the count for the baseline
comparison_counts['Baseline'] = len(baseline_df_top100)

# Set the x-axis labels
x_labels = rate_model_names + ['Baseline']

# Set the x-axis positions
x_pos = range(len(x_labels))

# Get the counts for each model
counts = [comparison_counts[model_name] for model_name in x_labels]

# Create the bar plot
plt.figure(figsize=(6, 6))
plt.bar(x_pos, counts)
plt.xlabel('Model')
plt.ylabel('Number of Users')
plt.title('Similar counts of Highest Impact Dimensions of \n Regression Models with Baseline (Per-User)')
plt.xticks(x_pos, x_labels, rotation=90)
plt.tight_layout()
plt.show()
```
